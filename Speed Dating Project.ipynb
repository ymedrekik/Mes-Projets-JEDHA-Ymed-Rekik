Attraction between Males and Females by Interests

```python
correl_m = df[df.gender=='Male'][interests].corr()
correl_f = df[df.gender=='Female'][interests].corr()

fig, ax= plt.subplots(1,2, figsize=(16, 8))
sns.heatmap(correl_m, cmap='magma',annot=True, ax=ax[0])
sns.heatmap(correl_f, cmap='magma',annot=True, ax=ax[1])

ax[0].set_title('Males', fontsize=12)
ax[1].set_title('Females', fontsize=10)
```

```python
g = sns.FacetGrid(df, col='race')
g.map(plt.hist, 'imprace', bins=12)
axes = g.axes.flatten()
axes[0].set_title("Asian")
axes[1].set_title("White")
axes[2].set_title("Other")
axes[3].set_title("Hispanic")
axes[4].set_title("Black")
g.set_xlabels('', fontsize=1)
plt.subplots_adjust(top=0.8)
g.fig.suptitle('How important is the Roots, by Roots', fontsize=18)
g.add_legend()
```
